```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt

#Set Display options
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', None)
```

```python
all_players = pd.read_csv(r'../data/2020-21/ML_dataset/ML_all_players.csv')
```

```python
all_players_sum = all_players.groupby(by='full_name').sum()
all_players_sum.sort_values(by= "xG" , ascending = False , inplace = True)
top_50 = all_players_sum[:50]
```

```python
top_50.plot(
    kind = 'scatter',
    x = "xG",
    y = "goals_scored",
    figsize = (10,6)
)
```

```python
def scatter_chart(data , x_label , y_label):

    x = data[x_label]
    y = data[y_label]

    plt.scatter(x, y , marker = 'o')
    plt.plot(y, y, color ='r')
    plt.title('Scatter Chart')
    plt.xlabel(y_label)
    plt.ylabel(x_label)
    #plt.legend()
    plt.show()
```

```python
scatter_chart(top_50 , 'xG' , 'goals_scored')
```

```python
scatter_chart(top_50, 'xA' , 'assists')
```
